```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image= cv2.imread('out.jpg')
```

```python
image_tt = image.copy()
lower_red = np.array([0,120,70])
upper_red = np.array([10,255,255])
image_tt = cv2.cvtColor(image_tt,cv2.COLOR_BGR2HSV)
mask1 = cv2.inRange(image_tt, lower_red, upper_red)
result_out = cv2.bitwise_and(image_tt,image_tt,mask = mask1)
result_out[:,:,0]=0
result_out[:,:,1]=0
plt.imshow(result_out)
plt.show()
```

```python
from numpy.random import seed
from numpy.random import randint
seed(1)
#ran=randint(0,3000,3000)
for i in np.arange(2000,4000):
    var = (i-2000)*10
    index = randint(0,6000,var)
    for j in np.arange(0,index.shape[0]):
        result_out[i,index[j],2]=0
```

```python
plt.imshow(result_out)
plt.show()
```

```python
rr = []
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
region=['A','B','C','D','E']
s=0
for u in np.arange(2200,4000):
    n=0
    for v in np.arange(0,result_out.shape[1]):
        if result_out[u,v,2] != 0:
            n=n+1
    uu = u%360
    if uu != 0  :
        s = s+n
    elif u!=2200 and uu==0:
        rr.append(s)
        s=0   
print(rr)
ax.bar(region,rr)
plt.show()
```

```python
result_copy = result_out.copy()
image_final = cv2.imread('focus.jpg')
n= 0
for i in np.arange(0,image_final.shape[0]):
    for j in np.arange(0,image_final.shape[1]):
        if result_copy[n,j,2]!= 0:
            image_final[i,j,:]=result_copy[n,j,:]
    n=n+1
plt.imshow(image_final)
plt.show()        
```

```python
cv2.imwrite('final_result.jpg',image_final)
```
